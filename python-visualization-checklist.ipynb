```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# load data
music = pd.read_csv('Music Sales.csv')
```

# 1. Bar chart

```python
list_of_values = ['Europe', 'Asia']
asia_euro = music[music['Region'].isin(list_of_values)]
asia_euro = asia_euro.groupby('Genre').sum().sort_values(by = 'Quantity', ascending = False)
asia_euro.head()
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(x="Quantity", y=asia_euro.index, data=asia_euro, hue = asia_euro.index, dodge = False)
```

# 2. Histogram

```python
plt.hist(music['Sales']);
```

```python
# using seaborn
# kde=True/False to turn on/off gaussian kernel density estimate plot
sns.distplot(music['Sales'], kde=True);
```

# 3. Box plot

```python
#sns boxplot
sns.set(style="ticks", palette="pastel")
sns.boxplot(x='Region', y="Sales", data=music)
```

# 4. Violin plot

```python
#sns boxplot
sns.set(style="ticks", palette="pastel")
sns.violinplot(x='Region', y="Sales", data=music)
```

# 5. scatter plot

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
#matplolib scatter plot
plt.scatter(iris["sepal_width"],iris["sepal_length"], alpha=0.8, marker=r'$\clubsuit$')
plt.xlabel("sepal_width")
plt.ylabel("sepal_length")
plt.legend(loc=2)
plt.show()
```

```python
# matplolib scatter plot
iris = sns.load_dataset("iris")
sns.pairplot(x_vars=["sepal_width"], y_vars=["sepal_length"], data=iris, hue="species", size=5)
```

# 9. Multiple Linear Regression

```python
iris = sns.load_dataset("iris")
sns.set()

# Load the iris dataset
# Plot sepal with as a function of sepal_length across days
g = sns.lmplot(x="sepal_length", y="sepal_width", hue="species",
               truncate=True, height=5, data=iris)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Sepal length (mm)", "Sepal width (mm)")
```
